```python
start_time = time.time()

# Define physical model
my_model = Model(material='dna', celsius=50)

# Set number of orthogonal systems
systems = len(a_toe)-1

# Initialize working variables
crosstalk_targets = {}
crosstalk_excludes = []
step_tubes = []


# Loop over orthogonal systems
for i in range(systems):
    
    # Define sequence domains
    da = Domain(a_toe[i], name=['da', i])
    db = Domain('N'*barcode_length, name=['db', i])
    dc = Domain(c_dup[i+1], name=['dc', i])

    # Define strands containing these domains
    sR1 = TargetStrand([da,db], name=['sR1', i])
    sR2 = TargetStrand([~db,~da], name=['sR2', i])
    sCtop = TargetStrand([~db,dc], name=['sCtop', i])
    sCbottom = TargetStrand([~dc, ~da], name=['sCbottom', i])

    
    # Define target complexes
    R1 = TargetComplex([sR1], '.'*(toehold_length+barcode_length), name=['R1', i])
    R2 = TargetComplex([sR1], '.'*(toehold_length+barcode_length), name=['R2', i])
    J1 = TargetComplex([sR1, sCtop, sCbottom],'('*(toehold_length+barcode_length)+'+'+')'*barcode_length+'('*len(c_dup[i+1])+'+'+')'*len(c_dup[i+1])+')'*toehold_length, name=['J1', i])
    C1 = TargetComplex([sCtop, sCbottom],'.'*barcode_length+'('*len(c_dup[i+1])+'+'+')'*len(c_dup[i+1])+'.'*toehold_length, name=['C1', i])
    
    
    
    # Define 2 elementary step tubes for each orthogonal system
    step0_tube = TargetTube(on_targets={R1: 1e-08, C1: 1e-08}, 
                           off_targets=SetSpec(max_size=2, exclude=[J1]), name=['reactants', i])
    step1_tube = TargetTube(on_targets={J1: 1e-08}, 
                          off_targets=SetSpec(max_size=2), name=['products', i])
    step_tubes += [step0_tube, step1_tube]
    
    # Store contributions to global crosstalk tube
    crosstalk_targets.update({R1: 1e-08, C1: 1e-08})
    crosstalk_excludes += [J1]
    
# Define 1 global crosstalk tube 
crosstalk_tube = TargetTube(on_targets=crosstalk_targets, 
                        off_targets=SetSpec(max_size=2, exclude=crosstalk_excludes), name='crosstalk')

# Define tube set    
my_tubes = step_tubes + [crosstalk_tube]    
    
# Define defect weight of N for global crosstalk tube 
# (maintain balance between elementary step tubes and global crosstalk tube as number of 
# orthogonal systems increases) 
my_weights = Weights(my_tubes)
my_weights[:,:,:,crosstalk_tube] *= systems

# Set a stop condition of 2% 
# Set seed for random number generation to get a reproducible result for this demo
my_options = DesignOptions(f_stop=0.10, seed=93)

# Define and run the test tube design job
my_design = tube_design(tubes=my_tubes, #hard_constraints=[Similarity([da, db], 'S25', limits=[0.35, 0.65]), Diversity(word=8, types=2)],
                        defect_weights=my_weights, model=my_model, options=my_options)
my_results = my_design.run(trials=1)[0]

# Display the design results
print("--- %s seconds ---" % (time.time() - start_time))
my_results
```
